Inicialmente devemos importar a biblioteca pandas que será usada para realizar essa análise dos dados


```python
import pandas as pd
import plotly.express as px
```

Após importar as bibliotecas, criamos um DataFrame para cada uma das páginas da planilha que está sendo usada.


```python
df_principal = pd.read_excel("Aula 2 - Imersão Python - Tabela de ações.xlsx", sheet_name="Principal")
df_principal
```

```python
df_total_acoes = pd.read_excel("Aula 2 - Imersão Python - Tabela de ações.xlsx", sheet_name="Total_de_acoes")
df_total_acoes
```

```python
df_ticker = pd.read_excel("Aula 2 - Imersão Python - Tabela de ações.xlsx", sheet_name="Ticker")
df_ticker
```

```python
df_chatgpt = pd.read_excel("Aula 2 - Imersão Python - Tabela de ações.xlsx", sheet_name="ChatGPT")
df_chatgpt
```

Porém só serão usadas as primeiras 4 colunas do DataFrame principal, com isso devemos descartar todas as outras.

Para fazer isso usamos a função **.copy()** que faz uma cópia das colunas que queremos e atribuimos isso ao DataFrame que está sendo usado 


```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

Como algumas colunas possuem nomes com caracteres especiais será necessário renomea-las para um padrão que diminua as chances de dar algum erro

Isso é feito usando a função **rename()** que possui o atributo columns onde são inseridos os nomes atuais das colunas seguido por dois pontos (:) e o novo nome da coluna

Além disso foi usado o copy() novamente para não nenhum erro

```python
df_principal = df_principal.rename(columns={'Último (R$)': 'Valor_final', 'Var. Dia (%)': 'Var_dia_pct'}).copy()
df_principal
```

## Criando novas colunas

Para criar novas colunas devemos somente colocar o nome do DataFrame e ente entre colchetes o nome dessa nova coluna. Ex: `nome_do_DataFrame['Nome_da_NovaColuna']`

**Var_pct:** A coluna var_pct equivale a porcentagem da coluna Var_dia_pct, por isso atribuimos a Var_pct a coluna Var_dia_pct dividida por 100

**Valor_inicial:** Para está coluna foi usada a fórmula *valor_final/(Variacao_porcentagem + 1)*

```python
df_principal['Var_pct'] = df_principal['Var_dia_pct'] / 100
df_principal['Valor_inicial'] = df_principal['Valor_final'] / (df_principal['Var_pct'] + 1)
df_principal
```

### Merge

No python não existe as funções PROCV ou VLOOKUP, então para realizar essas ações é ncessário usar a função **.merge()**

Essa função funciona atribuindo ao DataFrame(esquerada).merge(DataFrame(direita), left_on="Coluna_Esquerda", right_on="Coluna_Direita", how="Letf")

sendo coluna_esquerda a coluna do DataFrame(esquerda) que será usada como base para fazer a correspondencia dos valores

Coluna_direita é a coluna do DataFrame (direita) que é igual a coluna_esquerda pórem não está no DataFrame principal (esquerda)

left significa que tudo isso será atribuido ao DataFrame a esquerda

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

Como essa função junta dois DataFrames, a coluna que foi usada para fazer a correspondencia acaba sendo copiada, com isso devemos usar a função **.drop()** que receve como atributo columns as colunas que serão excluidas

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

Continuamos criando as próximas colunas
- Variação em reais

```python
df_principal['Variacao_rs'] = (df_principal['Valor_final'] - df_principal['Valor_inicial']) * df_principal['Qtde. Teórica']
```

### Formatando colunas

Como algumas colunas estão em notação cientifica devemos formata-las para o tipo float

```python
pd.options.display.float_format = '{:.2f}'.format
```

Mas a com isso acabamos formatando a coluna da quantidade teórica para float também, mas essas coluna só possui valores inteiros, e para voltarmos ela int novamente devemos utlizar

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal = df_principal.rename(columns={'Qtde. Teórica': 'Qtd_teorica'}).copy()
```

### Outras funções para criação de colunas

A função apply é utilizada para aplicar outras funções
A função labda serve para aplicar uma função linha a linha

```python
""" 
Outra forma de fazer está função condicional

def resultado(x):
    if x > 0:
        res = "Subiu"
        return res
    elif x < 0:
        res = "Desceu"
        return res
    else:
        res = "Estável"
        return res

df_principal['Resultado'] = df_principal['Variacao_rs'].apply(resultado)
"""

df_principal['Resultado'] = df_principal['Variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else('Desceu' if x < 0 else 'Estável'))
df_principal
```

Para terminar de completar o DataFrame fizemos as tabelas:
- Nome
- Segmento
- Idade
- Cat_idade

```python
df_principal = df_principal.merge(df_ticker, left_on="Ativo", right_on="Ticker", how='left')
df_principal = df_principal.drop(columns=['Ticker'])

df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Empresa', how='left')
df_principal = df_principal.drop(columns=['Empresa'])

df_principal['Cat_idade'] = df_principal['Idade'].apply(lambda x: 'Mais de 100' if x > 100 else('Menos de 50' if x < 50 else 'Entre 100 e 50'))

df_principal
```

## Análises da planilha

Inicialmente calculamos os maiores e menores valores da planilha, e após isso fizemos uma média desses valores, foram feitas as médias dos que subiram e dos que desceram para ter uma análise mais correta

```python
maior = df_principal['Variacao_rs'].max() # Função que pega o maior valor

menor = df_principal['Variacao_rs'].min() # Função que pega o menor valor

media = df_principal['Variacao_rs'].mean() # Função que faz a média doa valores

media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()

media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rs'].mean()

# Imprimindo os resultados
print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Média\tR$ {media:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Media de quem desceu\tR$ {media_desceu:,.2f}")
```

Após isso criamos um DataFrame que possui somente os valores que subiram

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

Continuamos criando um DataFrame que possui os valores de variação organizados por segmento

Para isso usamos a função groupby que recebeu como parametro a coluna que queriamos organizada, mas essa função retorna um vetor, portanto usamos a função .reset_index() para retornarmos para um DataFrame 

```python
df_analise_segmento = df_principal[df_principal['Resultado'] == 'Subiu'].groupby('Segmento')['Variacao_rs'].mean().reset_index()
df_analise_segmento
```

Continuamos criando um DataFrame para se ter uma análise da média dos resultados

```python
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rs'].mean().reset_index()
df_analise_saldo
```

## Criando Gráficos

Para podermos criar os gráficos utilizamos a biblioteca plotly.express por isso devemos importa-la junto com o pandas

Para fazermos um gráfico do DataFrame analise de saldo criamos uma variável chamada fig de figura, e atribuimos a ela a função **px.bar** que cria um gráfico de barras, e que recebe como atributos o DataFrame que será utilizado, qual coluna estará no eixo x e aul estará no eixo y, quais valores estarão escritos nas barras e um titulo o gráfico.

```python
fig = px.bar(df_analise_saldo, x= 'Resultado', y='Variacao_rs', text='Variacao_rs', title='Variação Reais por resultado')
fig.show()
```
